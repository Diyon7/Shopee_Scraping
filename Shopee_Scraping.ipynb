```python
def getUrl(a):

  title = []
  image = []
  price = []

  url = urllib.request.urlopen(a)
  result = url.read()
  url.close()

  soup = BeautifulSoup(result, "html.parser")

  for tle in soup.find_all('div', class_='ZG__4J'):
    title.append(tle.get_text())
  for img in soup.find_all('img', class_='_3-N5L6 _2GchKS'):
    image.append(img.get('src'))
  for pic in soup.find_all('div',class_='zp9xm9 kNGSLn l-u0xK'):
    price.append(pic.get_text())

  print(title)
  print(image)
  print(price)
  return image, title, price, soup
```

```python
image = pd.DataFrame(columns=['judul','link','tanggal'])
```

```python
kompas='https://shopee.co.id/Top-Handle-Bags-col.1080936?filters=6%2C7&page=0&sortBy=sales'
title, image, price, semua = getUrl(kompas)
for a in range(len(title)):
  image = image.append(pd.DataFrame([[title[a],image[a],price[a]]],columns=['judul','link','tanggal']),ignore_index=True)
```

```python
image.to_csv('integrasi_level_ui.csv',index=False)
```

```python
product_name, product_price, product_image = [], [], []

# product name
for name in soup.find_all('div', class_='ZG__4J') :
    product_name.append(name.get_text())

# product price
for price in soup.find_all('div', class_='zp9xm9 kNGSLn l-u0xK') :
    product_price.append(price.get_text())

# product image
for image in soup.find_all('img', class_='_3-N5L6 _2GchKS') :
    product_image.append(image.get('src'))
```

```python
product_name, product_price, product_image, product_link = [], [], [], []

for page in range(0,2):
    main_link='https://shopee.co.id/Top-Handle-Bags-col.1080936?filters=6%2C7&page={}&sortBy=sales'.format(page)
    driver.get(main_link)

    # product name
    for name in soup.find_all('div', class_='ZG__4J') :
        product_name.append(name.get_text())

    # product price
    for price in soup.find_all('div', class_='zp9xm9 kNGSLn l-u0xK') :
        product_price.append(price.get_text())

    # product image
    for image in soup.find_all('img', class_='_3-N5L6 _2GchKS') :
        product_image.append(image.get('src'))

    # product link
    for link in soup.find_all('div', class_='col-xs-2-4 shopee-search-item-result__item') :
        for linkhref in link.find_all('a') :
          product_link.append(linkhref.get('href'))
```

```python
listCols = ['product_name','product_price', 'product_image']
dict_data = dict(zip(
            listCols,
                (product_name,
                 product_price,
                 product_link)
))
```
